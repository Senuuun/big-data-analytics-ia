```python
notas = []

for i in range(5):
    nota = float(input(f"Digite a nota do aluno {i + 1}: "))
    notas.append(nota)

media = sum(notas) / len(notas)

if notas[0] >= media:
    situacao = "APROVADO"
else:
    situacao = "REPROVADO"

print(f"\nMédia da Turma: {media}")
print(f"Portanto, o aluno 1 está {situacao}!")
```
