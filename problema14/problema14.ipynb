```python
import pandas as pd

colunas = [
    "Classe", "Alcool", "Acido_Malico", "Cinzas",
    "Alcalinidade_Cinzas", "Magnesio", "Fenóis_Totais",
    "Flavanoides", "Fenóis_Nao_Flavanoides",
    "Proantocianinas", "Intensidade_Cor",
    "Matiz", "OD280_OD315", "Prolina"
]

dados = pd.read_csv("wine.data", header=None, names=colunas)

print("Quantidade de Vinhos por Tipo:")
print(dados["Classe"].value_counts().sort_index())

print("\nMaior teor alcoólico:")
print(dados["Alcool"].max())

print("\nMenor teor alcoólico:")
print(dados["Alcool"].min())

print("\nMédia de Teor Alcoólico por Tipo de Vinho:")
print(dados.groupby("Classe")["Alcool"].mean())
```
