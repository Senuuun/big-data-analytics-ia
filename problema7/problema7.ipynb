```python
jogos = {}

for i in range(1, 6):
    print(f"\nJogo {i}")

    gols_avai = int(input("Digite os gols do Avaí: "))
    gols_figueira = int(input("Digite os gols do Figueirense: "))

    jogos[i] = {
        "avai": gols_avai,
        "figueirense": gols_figueira
    }

vitorias_avai = 0
vitorias_figueira = 0
empates = 0
saldo_avai = 0
saldo_figueira = 0

print("\nListagem de jogos:")

for jogo, resultado in jogos.items():
    avai = resultado["avai"]
    figueira = resultado["figueirense"]

    print(f"Jogo {jogo} (Avaí x Figueirense): {avai} x {figueira}")

    if avai > figueira:
        vitorias_avai += 1
    elif figueira > avai:
        vitorias_figueira += 1
    else:
        empates += 1

    saldo_avai += avai - figueira
    saldo_figueira += figueira - avai

print("\nResultados:")
print(f"Quantidade de vitórias do Avaí: {vitorias_avai}")
print(f"Quantidade de vitórias do Figueirense: {vitorias_figueira}")
print(f"Quantidade de empates: {empates}")
print(f"Saldo de gols do Avaí: {saldo_avai}")
print(f"Saldo de gols do Figueirense: {saldo_figueira}")
```
