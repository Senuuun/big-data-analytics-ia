```python
emails = set()

with open(r"C:\Users\arthur_c_nunes\Downloads\mailbox.txt", "r") as arquivo:
    for linha in arquivo:

        if linha.startswith("From"):
            partes = linha.split()
            email = partes[1]
            emails.add(email)

with open("output.txt", "w") as saida:
    for email in emails:
        saida.write(email + "\n")

print("Arquivo output.txt gerado com sucesso!")
```
