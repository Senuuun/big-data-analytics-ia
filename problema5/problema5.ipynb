```python
def melhor_combustivel(gasolina, etanol):
    if gasolina < 0.01 or gasolina > 10 or etanol < 0.01 or etanol > 10:
        return "PREÇO INVÁLIDO"
    
    relacao = etanol / gasolina

    if relacao <= 0.73:
        return "ETANOL"
    else:
        return "GASOLINA"

print("Bem vindo a Calculadora do melhor combustível!")

gasolina = float(input("Digite o preço da Gasolina em R$: "))
etanol = float(input("Digite o preço do Etanol em R$: "))

resultado = melhor_combustivel(gasolina, etanol)

print(f"Para os preços informados, a melhor opção é a utilização da {resultado}")
```
