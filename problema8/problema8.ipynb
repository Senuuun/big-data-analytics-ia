```python
temperaturas = {}

meses = [
    "janeiro", "fevereiro", "março", "abril",
    "maio", "junho", "julho", "agosto",
    "setembro", "outubro", "novembro", "dezembro"
]

for mes in meses:
    temp = float(input(f"Digite a temperatura média de {mes}: "))
    temperaturas[mes] = temp

print("\n-- Temperaturas do Ano ---")

for mes, temp in temperaturas.items():
    print(f"{mes}: {temp}°C")

maior_temp = max(temperaturas.values())
menor_temp = min(temperaturas.values())

for mes, temp in temperaturas.items():

    if temp == maior_temp:
        mes_maior = mes

    if temp == menor_temp:
        mes_menor = mes

print(f"\nMaior temperatura do ano: {maior_temp}°C em {mes_maior}")
print(f"Menor temperatura do ano: {menor_temp}°C em {mes_menor}")
```
