```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1.1 - Ler o dataset
df = pd.read_csv("movie_metadata.csv")

# 1.2 - Valores faltantes
faltantes_director = df["director_name"].isna().sum()
faltantes_duration = df["duration"].isna().sum()

print("Valores faltantes em director_name:", faltantes_director)
print("Valores faltantes em duration:", faltantes_duration)

# 1.3 - Substituir valores ausentes de duration pela média
media_duration = df["duration"].mean()
df["duration"] = df["duration"].fillna(media_duration)

print("Média da duration usada para preencher:", media_duration)

# 1.4 - Média do imdb_score
media_imdb_score = round(df["imdb_score"].mean(), 2)

print("Média do imdb_score:", media_imdb_score)

# 1.5 - Criar coluna movie_duration_category
condicoes = [
    df["duration"] < 90,
    df["duration"].between(90, 120, inclusive="both"),
    df["duration"] > 120
]

categorias = [
    "Categoria 1",
    "Categoria 2",
    "Categoria 3"
]

df["movie_duration_category"] = np.select(
    condicoes,
    categorias,
    default="Sem categoria"
)

print(df[["duration", "movie_duration_category"]].head())

# 1.6 - Filmes entre 90 minutos e 2 horas lançados em 2008
filmes_2008_categoria_2 = df.query(
    'title_year == 2008 and movie_duration_category == "Categoria 2"'
).shape[0]

print("Filmes entre 90 minutos e 2 horas lançados em 2008:", filmes_2008_categoria_2)

# 1.7 - Média do imdb_score por país, top 15
media_por_pais = (
    df.groupby("country")["imdb_score"]
    .mean()
    .round(2)
    .sort_values(ascending=False)
    .head(15)
)

print("\nTop 15 países por média de imdb_score:")
print(media_por_pais)

# 1.8 - Quantidade de filmes por país, top 15
quantidade_por_pais = df["country"].value_counts().head(15)

print("\nTop 15 países por quantidade de filmes:")
print(quantidade_por_pais)

# Gráfico de barras
plt.figure(figsize=(18, 6))

barras = plt.bar(
    quantidade_por_pais.index,
    quantidade_por_pais.values,
    alpha=0.7
)

plt.title("Quantidade de Filmes")
plt.xlabel("country")
plt.ylabel("Quantidade de Filmes")
plt.grid(True)

# Mostrar os valores em cima das barras
for barra in barras:
    altura = barra.get_height()
    plt.text(
        barra.get_x() + barra.get_width() / 2,
        altura,
        int(altura),
        ha="center",
        va="bottom",
        fontsize=8
    )

plt.show()
```
