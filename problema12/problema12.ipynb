```python
import numpy as np

gols_avai = np.array([], dtype=int)
gols_figueira = np.array([], dtype=int)

for jogo in range(1, 6):

    partida = input(f"Digite o resultado do jogo {jogo} (Avaí x Figueirense): ")

    placar1, placar2 = partida.split("x")

    placar1 = int(placar1)
    placar2 = int(placar2)

    gols_avai = np.append(gols_avai, placar1)
    gols_figueira = np.append(gols_figueira, placar2)

print()

for i in range(5):
    print(f"Jogo {i+1}(Avaí x Figueirense):{gols_avai[i]}x{gols_figueira[i]}")

vitorias_avai = np.sum(gols_avai > gols_figueira)

vitorias_figueira = np.sum(gols_figueira > gols_avai)

empates = np.sum(gols_avai == gols_figueira)

saldo_avai = np.sum(gols_avai - gols_figueira)

saldo_figueira = np.sum(gols_figueira - gols_avai)

print(f"Vitórias do Avaí: {vitorias_avai}")
print(f"Vitórias do Figueirense: {vitorias_figueira}")
print(f"Empates: {empates}")
print(f"Saldo de gols do Avaí: {saldo_avai}")
print(f"Saldo de gols do Figueirense: {saldo_figueira}")
```
