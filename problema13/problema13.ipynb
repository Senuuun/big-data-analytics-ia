```python
import numpy as np

dados = np.loadtxt("wine.data", delimiter=",")

classes = dados[:, 0]

alcool = dados[:, 1]

classe1 = np.sum(classes == 1)
classe2 = np.sum(classes == 2)
classe3 = np.sum(classes == 3)

maior_alcool = np.max(alcool)

menor_alcool = np.min(alcool)

media_classe1 = np.mean(alcool[classes == 1])
media_classe2 = np.mean(alcool[classes == 2])
media_classe3 = np.mean(alcool[classes == 3])

print("Quantidade de Vinhos por Classe:")
print(f"Classe 1: {classe1}")
print(f"Classe 2: {classe2}")
print(f"Classe 3: {classe3}")

print(f"\nMaior teor alcoólico: {maior_alcool}")

print(f"Menor teor alcoólico: {menor_alcool}")

print("\nMédia de teor alcoólico por Tipo de Vinho:")
print(f"Classe 1: {media_classe1:.2f}")
print(f"Classe 2: {media_classe2:.2f}")
print(f"Classe 3: {media_classe3:.2f}")
```
