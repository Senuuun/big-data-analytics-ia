```python
import random

numeros = [random.randint(1, 100) for i in range(15)]

media = sum(numeros) / len(numeros)

quadrados = [(numero - media) ** 2 for numero in numeros]

print("Números aleatórios:", numeros)
print(f"Média: {media}")

print("Quadrados da diferença em relação à média:", quadrados)
```
