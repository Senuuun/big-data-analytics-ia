```python
jogos = {}

for jogo in range(1, 6):

    partida = input(f"Digite o resultado do jogo {jogo}: ")

    placar1, placar2 = partida.split("x")

    placar1 = int(placar1)
    placar2 = int(placar2)

    jogos[jogo] = {
        "avai": placar1,
        "figueirense": placar2
    }

vitorias_avai = 0
vitorias_figueira = 0
empates = 0
saldo_avai = 0
saldo_figueira = 0

print("\n----Resultados dos Jogos----")

for jogo, resultado in jogos.items():

    avai = resultado["avai"]
    figueira = resultado["figueirense"]

    print(f"Jogo {jogo} (Avaí x Figueirense): {avai}x{figueira}")

    if avai > figueira:
        vitorias_avai += 1

    elif figueira > avai:
        vitorias_figueira += 1

    else:
        empates += 1

    saldo_avai += avai - figueira
    saldo_figueira += figueira - avai

print(f"\nQuantidade de vitórias do Avaí: {vitorias_avai}")
print(f"Quantidade de vitórias do Figueirense: {vitorias_figueira}")
print(f"Quantidade de empates: {empates}")
print(f"Saldo de gols do Avaí: {saldo_avai}")
print(f"Saldo de gols do Figueirense: {saldo_figueira}")
```
